```python
data = [
    (1, "t1", "2024-01-01 10:00:00", 100.0),
    (1, "t2", "2024-01-01 15:30:00", 50.0),
    (1, "t3", "2024-01-02 09:00:00", 200.0),
    (2, "t4", "2024-01-01 11:00:00", 300.0),
    (2, "t5", "2024-01-02 14:00:00", 150.0),
    (2, "t6", "2024-01-02 18:00:00", 50.0)
]
```

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType, TimestampType, DoubleType
from datetime import datetime

# Convert transaction_ts from string to datetime
fixed_data = [
    (user_id, transaction_id, datetime.strptime(transaction_ts, "%Y-%m-%d %H:%M:%S"), amount)
    for user_id, transaction_id, transaction_ts, amount in data
]

schema = StructType([
    StructField("user_id", LongType(), False),
    StructField("transaction_id", StringType(), False),
    StructField("transaction_ts", TimestampType(), False),
    StructField("amount", DoubleType(), False),
])

df = spark.createDataFrame(fixed_data, schema)
df.show()
```

**Expected Output :**

| user_id | transaction_date | total_amount |
| ------- | ---------------- | ------------ |
| 1       | 2024-01-01       | 150.0        |
| 1       | 2024-01-02       | 200.0        |
| 2       | 2024-01-01       | 300.0        |
| 2       | 2024-01-02       | 200.0        |
