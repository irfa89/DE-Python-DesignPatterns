Problem: Daily Metrics Pivot

You are given a list of event records generated by an application.
Each record represents a metric value recorded for a user on a given day.

Your task is to pivot this data so that each (date, user_id) becomes a single record with metrics as columns.

Input Format
events = [
    ("2025-01-01", "u1", "clicks", 3),
    ("2025-01-01", "u1", "views", 10),
    ("2025-01-01", "u2", "clicks", 1),
    ("2025-01-02", "u1", "views", 7),
    ("2025-01-02", "u2", "clicks", 2),
]


Each tuple represents:

(date, user_id, metric_name, metric_value)

Output Format

Return a dictionary where:

{
  (date, user_id): {
      metric_name: aggregated_value
  }
}


Expected output for the above input:

{
  ("2025-01-01", "u1"): {"clicks": 3, "views": 10},
  ("2025-01-01", "u2"): {"clicks": 1},
  ("2025-01-02", "u1"): {"views": 7},
  ("2025-01-02", "u2"): {"clicks": 2}
}

Constraints

Assume input fits in memory

Same (date, user_id, metric_name) may appear multiple times

Metric values should be summed

Use Python dictionaries only (no pandas)

Problem: Country × Device Pivot (Session Counts)

You are given raw session logs from a website.
Each log records one session with the user’s country and device type.

Your task is to pivot the data so that for each country, you count sessions per device.

Input Format
sessions = [
    ("US", "mobile"),
    ("US", "desktop"),
    ("US", "mobile"),
    ("IN", "mobile"),
    ("IN", "desktop"),
    ("IN", "desktop"),
]


Each tuple represents:

(country, device_type)

Output Format

Return a dictionary where:

{
  country: {
      device_type: session_count
  }
}


Expected output:

{
  "US": {"mobile": 2, "desktop": 1},
  "IN": {"mobile": 1, "desktop": 2}
}

Constraints

Input size can be large

Order does NOT matter

Devices are not fixed (new device types may appear)

Use Python dictionaries / defaultdict only

Problem: Feature Usage Pivot per User

You are given application usage logs.
Each record represents how many times a feature was used by a user.

Your task is to pivot the data so that each user has a single record with feature usage counts.

Input Format
usage = [
    ("u1", "search", 5),
    ("u1", "filter", 2),
    ("u2", "search", 3),
    ("u1", "search", 1),
    ("u2", "export", 4),
]


Each tuple represents:

(user_id, feature_name, count)

Output Format

Return a dictionary:

{
  user_id: {
      feature_name: total_count
  }
}


Expected output:

{
  "u1": {"search": 6, "filter": 2},
  "u2": {"search": 3, "export": 4}
}

Constraints

Same (user_id, feature_name) can appear multiple times

Counts must be summed

Features are dynamic (unknown ahead of time)

Use Python dict / defaultdict only

Problem: Status Code Pivot (API Logs)

You are given API request logs.
Each record represents one request with an HTTP status code.

Your task is to pivot the data so that for each API endpoint, you count requests by status code.

Input Format
logs = [
    ("/login", 200),
    ("/login", 401),
    ("/login", 200),
    ("/orders", 200),
    ("/orders", 500),
    ("/orders", 200),
]


Each tuple represents:

(endpoint, status_code)

Output Format

Return a dictionary:

{
  endpoint: {
      status_code: count
  }
}


Expected output:

{
  "/login": {200: 2, 401: 1},
  "/orders": {200: 2, 500: 1}
}

Constraints

Status codes are integers

Endpoints are dynamic

Order does not matter

Use dictionaries / defaultdict only

Assume in-memory processing
