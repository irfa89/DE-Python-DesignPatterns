```python
sales = [
    ("p1", 3),
    ("p2", 5),
    ("p1", 2),
    ("p3", 7),
    ("p2", 1)
]
```

```python
from collections import defaultdict

agg = defaultdict(int)

for key,val in sales:
    agg[key] += val
agg
```

```python
transactions = [
    ("p1", 2, 20.0),
    ("p2", 1, 15.0),
    ("p1", 3, 30.0),
    ("p2", 4, 60.0),
    ("p3", 1, 10.0)
]
```

```python
from collections import defaultdict

product_agg = defaultdict(lambda:{"quantity":0,"revenue":0.0})

for p,q,r in transactions:
    product_agg[p]['quantity'] += q
    product_agg[p]['revenue']  += r

product_agg
```
