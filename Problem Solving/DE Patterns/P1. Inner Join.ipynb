```python

## Dimension table:
dim = [
  {"id":1, "name":"Alice"},
  {"id":2, "name":"Bob"}
]

## Fact table:
fact = [
  {"id":1, "amt":100},
  {"id":2, "amt":200},
  {"id":3, "amt":300}
]
```

```python
dim_index = {d["id"]: d for d in dim}
dim_index
```

```python
enriched = []
for f in fact:
    if f["id"] in dim_index:
        enriched.append({**f,**dim_index[f["id"]]})
enriched
```

```python
orders = [
    {"order_id": 1, "cust_id": "A"},
    {"order_id": 2, "cust_id": "B"},
    {"order_id": 3, "cust_id": "C"},
]

customers = [
    {"cust_id": "A", "name": "Alice"},
    {"cust_id": "B", "name": "Bob"},
]
```

```python
cust_idx = {c["cust_id"]: c for c in customers}
cust_idx
```

```python
joined = []
for o in orders:
    if o["cust_id"] in cust_idx:
        joined.append({**o,**cust_idx[o["cust_id"]]})
joined
```

```python
## users.json:
users = [ {"uid":1,"country":"US"}, {"uid":2,"country":"CA"} ]
## events.json:
events = [ {"uid":1,"ev":"open"}, {"uid":3,"ev":"click"}, {"uid":2,"ev":"open"} ]
```

```python
user_idx = {u["uid"]: u for u in users}
user_idx
```

```python
enriched = []
for e in events:
    if e["uid"] in user_idx:
        enriched.append({**e, **user_idx[e["uid"]]})
enriched
```
