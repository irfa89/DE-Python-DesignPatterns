**1. Problem Statement**

You are given two lists of dictionaries representing two tables: users and orders. You need to perform an inner join on user_id and return a list of dictionaries where each dictionary contains the combined data of matching users and orders.

This simulates a simple ETL transformation where you need to merge related datasets in memory.

**Input Format**
users = [
    {"user_id": 1, "name": "Alice"},
    {"user_id": 2, "name": "Bob"},
    {"user_id": 3, "name": "Charlie"}
]

orders = [
    {"order_id": 101, "user_id": 1, "amount": 250},
    {"order_id": 102, "user_id": 1, "amount": 450},
    {"order_id": 103, "user_id": 2, "amount": 300},
    {"order_id": 104, "user_id": 4, "amount": 150}
]

**Output Format**

Return a list of dictionaries with combined data for matching user_id. For the example above:

[
    {"user_id": 1, "name": "Alice", "order_id": 101, "amount": 250},
    {"user_id": 1, "name": "Alice", "order_id": 102, "amount": 450},
    {"user_id": 2, "name": "Bob", "order_id": 103, "amount": 300}
]

**Constraints**

Both input lists can have up to 10^5 entries.

user_id is unique in users but may repeat in orders.

Do not use any external libraries like pandas; only Python built-ins.

Aim for O(N + M) time complexity using a hash map.

```python
users = [
    {"user_id": 1, "name": "Alice"},
    {"user_id": 2, "name": "Bob"},
    {"user_id": 3, "name": "Charlie"}
]

orders = [
    {"order_id": 101, "user_id": 1, "amount": 250},
    {"order_id": 102, "user_id": 1, "amount": 450},
    {"order_id": 103, "user_id": 2, "amount": 300},
    {"order_id": 104, "user_id": 4, "amount": 150}
]
```

```python
hash_map = {}

for u in users:
    hash_map[u["user_id"]] = u["name"]

res = []
for o in orders:
    if o["user_id"] in hash_map:
        res.append({
            "user_id" : o["user_id"],
            "name" : hash_map[o["user_id"]],
            "order_id" : o["order_id"],
            "amount" : o["amount"]
        })
res
```
