Problem

Implement a circular queue using deque with max capacity k.

```python
cq = CircularQueue(2)
print(cq.enqueue(1))  # True
print(cq.enqueue(2))  # True
print(cq.enqueue(3))  # False
print(cq.dequeue())   # 1
print(cq.enqueue(3))  # True
```
