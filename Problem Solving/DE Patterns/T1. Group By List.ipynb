```python
rows = [
    {"cust": "a", "amt": 10},
    {"cust": "b", "amt": 20},
    {"cust": "a", "amt": 15}
]
```

```python
from collections import defaultdict

groups = defaultdict(list)

for row in rows:
    key = row["cust"]
    groups[key].append(row)
groups
```

```python
orders = [
    ("u1", "o101"),
    ("u2", "o201"),
    ("u1", "o102"),
    ("u2", "o202"),
    ("u1", "o103"),
]

from collections import defaultdict

groups = defaultdict(list)

for user, trans in orders:
    groups[user].append(trans)

groups
```

```python
events = [
    {"session_id": "s1", "event": "page_view", "timestamp": 100},
    {"session_id": "s1", "event": "click",     "timestamp": 105},
    {"session_id": "s2", "event": "page_view", "timestamp": 102},
    {"session_id": "s1", "event": "scroll",    "timestamp": 110},
    {"session_id": "s2", "event": "click",     "timestamp": 108},
    {"session_id": "s2", "event": "purchase",  "timestamp": 115},
]
```

```python
events_sorted = sorted(events, key = lambda x: (x["session_id"],x["timestamp"]))
```

```python
from collections import defaultdict

groups = defaultdict(list)

for e in events_sorted:
    groups[e["session_id"]].append(e)
groups
```

```python
sessions = defaultdict(list)

for e in events_sorted:
    sessions[e["session_id"]].append({
        "event": e["event"],
        "timestamp": e["timestamp"]
    })
sessions
```

```python
transactions = [
    {"account_id": "A1", "amount": 100, "timestamp": 170},
    {"account_id": "A2", "amount": 250, "timestamp": 172},
    {"account_id": "A1", "amount": 75,  "timestamp": 175},
    {"account_id": "A3", "amount": 400, "timestamp": 176},
    {"account_id": "A2", "amount": 125, "timestamp": 178},
]
```

```python
accnt_amts = defaultdict(list)

for t in transactions:
    accnt_amts[t["account_id"]].append(t["amount"])

accnt_amts
```

```python
logs = [
    {"service_name": "auth-service", "level": "INFO",  "message": "Login success"},
    {"service_name": "payment-service", "level": "ERROR", "message": "Card declined"},
    {"service_name": "auth-service", "level": "WARN",  "message": "Password retry"},
    {"service_name": "order-service", "level": "INFO",  "message": "Order created"},
    {"service_name": "payment-service", "level": "INFO",  "message": "Payment retry"},
]

logs_by_service = defaultdict(list)

for log in logs:
    logs_by_service[log["service_name"]].append(log)

logs_by_service
```

```python
logs_by_service = defaultdict(list)

for log in logs:
    logs_by_service[log["service_name"]].append(log["message"])

logs_by_service
```
