```python
from collections import defaultdict
groups = defaultdict(dict)

for u,p in events:
    groups[u][p] = groups[u].get(p,0) + 1
groups
```

```python
logs = [
    ("auth", 200),
    ("auth", 500),
    ("billing", 200),
    ("auth", 200),
    ("billing", 404),
    ("billing", 200),
    ("auth", 403),
    ("billing", 500),
]
```

```python
from collections import defaultdict

service_freq = defaultdict(dict)

for service, status in logs:
    service_freq[service][status] = service_freq[service].get(status,0) + 1

service_freq
```

```python
result = {}
for service,freq_dict in service_freq.items():
    sorted_list =  sorted(freq_dict.items(), key = lambda x:x[1], reverse = True)
    result[service] = sorted_list

result
```
