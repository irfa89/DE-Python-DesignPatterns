```python
users = [
    {"user_id": 1, "name": "Alice"},
    {"user_id": 2, "name": "Bob"},
    {"user_id": 3, "name": "Charlie"}
]

events = [
    {"user_id": 1, "event": "login"},
    {"user_id": 1, "event": "purchase"},
    {"user_id": 2, "event": "logout"},
    {"user_id": 4, "event": "login"}
]
```

```python
user_map = {u["user_id"]: u for u in users}
user_map
```

```python
user_map = {}
for u in users:
    user_map[u["user_id"]] = u["name"]
user_map
```

```python
result = []
for e in events:
    uid = e["user_id"]
    if uid in user_map:
        result.append({
            "user_id": uid,
            "name" : user_map[uid],
            "event" : e["event"]
        })
result
```
