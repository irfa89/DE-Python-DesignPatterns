**Problem Statement**

You are given a list of sales transactions. Each transaction is represented as a tuple containing (store_id, item_id, quantity_sold, revenue). Your task is to aggregate the total quantity_sold and revenue for each store_id and return the result as a dictionary.

**Input Format**
transactions = [
    (1, "apple", 10, 20.0),
    (2, "banana", 5, 7.5),
    (1, "orange", 8, 16.0),
    (2, "apple", 3, 6.0),
    (1, "banana", 2, 4.0)
]

store_id: integer

item_id: string

quantity_sold: integer

revenue: float

**Output Format**

A dictionary where each key is store_id and the value is another dictionary containing the aggregated totals for quantity_sold and revenue.

Example Output

{
    1: {"quantity_sold": 20, "revenue": 40.0},
    2: {"quantity_sold": 8, "revenue": 13.5}
}

**Constraints**

store_id and item_id are guaranteed to be non-empty.

quantity_sold and revenue are always non-negative.

The list of transactions can have up to 10^5 entries.

```python
transactions = [
    (1, "apple", 10, 20.0),
    (2, "banana", 5, 7.5),
    (1, "orange", 8, 16.0),
    (2, "apple", 3, 6.0),
    (1, "banana", 2, 4.0)
]

from collections import defaultdict

tx_agg = defaultdict(lambda : {'quantity_sold':0,'revenue':0})

for store_id,item_id,quantity_sold,revenue in transactions:
    tx_agg[store_id]['quantity_sold'] += quantity_sold
    tx_agg[store_id]['revenue'] += revenue

tx_agg
```
