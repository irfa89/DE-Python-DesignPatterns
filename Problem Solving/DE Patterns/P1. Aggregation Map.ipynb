```python
rows = [
  {"cust":"a", "amt": 10},
  {"cust":"b", "amt": 20},
  {"cust":"a", "amt": 5}
]
```

```python
agg = {}

for row in rows:
    key = row["cust"]
    val = row["amt"]
    agg[key] = agg.get(key,0) + val
agg
```

```python
transactions = [
    ("A", 100),
    ("B", 50),
    ("A", 25),
    ("B", 75),
    ("C", 20),
]
```

```python
from collections import defaultdict

revenue = defaultdict(int)

for user,amount in transactions:
    revenue[user] += revenue.get(user,0) + amount

revenue
```
