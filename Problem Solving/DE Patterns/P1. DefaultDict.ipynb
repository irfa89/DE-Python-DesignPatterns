Easy 1 — Group values by key (Basic Group By)
Problem Statement

Given a list of (key, value) pairs, group all values for the same key.

Input
pairs = [("a", 1), ("b", 2), ("a", 3), ("b", 4)]

Output
{
  "a": [1, 3],
  "b": [2, 4]
}

Easy 2 — Count events per user
Problem Statement

Given a list of user IDs, count how many times each user appears.

Input
users = ["u1", "u2", "u1", "u3", "u2", "u1"]

Output
{
  "u1": 3,
  "u2": 2,
  "u3": 1
}

Medium 1 — Multi-level grouping (country → city → users)
Input
records = [
  ("US", "NY", "u1"),
  ("US", "NY", "u2"),
  ("US", "SF", "u3"),
  ("IN", "BLR", "u4")
]

Output
{
  "US": {
    "NY": ["u1", "u2"],
    "SF": ["u3"]
  },
  "IN": {
    "BLR": ["u4"]
  }
}
