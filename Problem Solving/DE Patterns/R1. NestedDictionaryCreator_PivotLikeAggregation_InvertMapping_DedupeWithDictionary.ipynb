âœ… EASY â€” Problem 1 of 6
Nested Dictionary Creator (Easy)
Problem Statement

You are given a list of user activity events.
Each event contains a user_id, event_type, and count.

Build a nested dictionary structured as:

{
  user_id: {
     event_type: total_count
  }
}


If the same (user_id, event_type) appears multiple times, sum the counts.

Input Format
events = [
    ("u1", "click", 2),
    ("u1", "view", 5),
    ("u2", "click", 1),
    ("u1", "click", 3),
    ("u2", "view", 4)
]

Output Format
{
    "u1": {
        "click": 5,
        "view": 5
    },
    "u2": {
        "click": 1,
        "view": 4
    }
}

Constraints

1 <= len(events) <= 10^5

Counts are positive integers

Input fits in memory

Order does not matter

```python
events = [
    ("u1", "click", 2),
    ("u1", "view", 5),
    ("u2", "click", 1),
    ("u1", "click", 3),
    ("u2", "view", 4)
]
```

```python
from collections import defaultdict

result = defaultdict(lambda:defaultdict(int))

for user_id,event_type,total_count in events:
    result[user_id][event_type] += total_count
final_result = {user_id: dict(events_dict) for user_id, events_dict in result.items()}
final_result
```

Problem

Given a list of (country, city, population) tuples, build a nested dictionary:

{
  country: {
     city: population
  }
}

Input
records = [
  ("US", "NYC", 8),
  ("US", "SF", 1),
  ("IN", "BLR", 12)
]

Output
{
  "US": {"NYC": 8, "SF": 1},
  "IN": {"BLR": 12}
}

```python
records = [
  ("US", "NYC", 8),
  ("US", "SF", 1),
  ("IN", "BLR", 12)
]
```

```python
from collections import defaultdict

result = defaultdict(lambda: defaultdict(int))

for country,city,pop in records:
    result[country][city] = pop
final_result = {country:dict(city) for country,city in result.items()}
final_result
```

```python
result = {}

for country, city, pop in records:
    if country not in result:
        result[country] = {}
    result[country][city] = pop
```
