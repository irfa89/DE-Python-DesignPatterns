**Fixed Window**

```python
nums1 = [2, 1, 5, 1, 3, 2]
k1 = 3
nums2 = [2, 3, 4, 1, 5]
k2 = 2

def max_sum_subarr(nums,k):
    l, window_sum , max_sum = 0, 0, float('-inf')

    for r in range(len(nums)):
        window_sum += nums[r]

        if r - l + 1 == k:
            max_sum = max(max_sum,window_sum)
            window_sum -= nums[l]
            l += 1
    return max_sum

print(max_sum_subarr(nums1,k1))
print(max_sum_subarr(nums2,k2))
```

```python
nums1 = [5, -2, 3, 4]
k1 = 3
nums2 = [8, -1, 3, 4]
k2 = 2

def max_sum_circ_arr(nums,k):
    l , n  = 0 , len(nums)
    window_sum = sum(nums[i%n] for i in range(k))
    max_sum = window_sum
    for i in range(n):
        window_sum += nums[ (i + k - 1) % n]
        window_sum -= nums[ (i - 1) % n]
        max_sum = max(max_sum,window_sum)
    return max_sum
print(max_sum_circ_arr(nums1,k1))
print(max_sum_circ_arr(nums2,k2))
```

```python
nums1 = [5, -2, 3, 4]
k1 = 3
nums2 = [8, -1, 3, 4]
k2 = 2

def max_sum_circ_arr(nums,k):
    n = len(nums)
    window_sum = sum(nums[i%n] for i in range(k))
    max_sum = window_sum
    for i in range(1,n):
        window_sum += nums[(i + k - 1)%n]
        window_sum -= nums[(i - 1)% n]
        max_sum = max(max_sum, window_sum)
    return max_sum

print(max_sum_circ_arr(nums1,k1))
print(max_sum_circ_arr(nums2,k2))
```

**Variable Window (At Most)**

```python
s1 = "eceba"
k1 = 2

s2 = "aa"
k2 = 1

def max_substr_k_distinct_char(s,k):
  l, freq, max_len = 0 , {}, 0

  for r in range(len(s)):
    freq[s[r]] = freq.get(s[r],0) + 1
    while len(freq) > k:
      if freq[s[r]] > 1:
        freq[s[r]] = s[l]
        del freq[s[l]]
        l += 1
      max_len = max(max_len,r - l + 1)
  return max_len 

print(max_substr_k_distinct_char(s1,k1))
```
