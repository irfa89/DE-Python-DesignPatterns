```python
nums1 = [1,1,1]
k1 = 2 # Output: 2
nums2 = [1,2,3]
k2 = 3 #Output: 2

# current_sum - earlier_sum = k ; earlier_sum = current_sum - k
def subarr_sum(nums,k):
    prefix = {0:1}
    curr = 0
    count = 0

    for x in nums:
        curr += x
        if curr - k in prefix:
            count += prefix[curr - k]
        prefix[curr] = prefix.get(curr,0) + 1
    return count
        
print(subarr_sum(nums1,k1))
print(subarr_sum(nums2,k2))
```

```python
s1 = "leetcode"
s2 = "aabb"

def first_non_repeat(s):
    freq = {}
    for x in s:
        freq[x] = freq.get(x,0) + 1

    for i,x in enumerate(s):
        if freq[x] == 1:
            return i
        else:
            return -1
print(first_non_repeat(s1))
print(first_non_repeat(s2))
```

```python
nums1 = [1,1,1,2,2,3]
k1 = 2

nums2 = [4,4,4,6,6,7]
k2 = 1

def top_most_freq(nums, k):
    freq = {}
    for n in nums:
        freq[n] = freq.get(n, 0) + 1
    return [num for num, _ in sorted(freq.items(), key=lambda x: x[1], reverse=True)[:k]]

print(top_most_freq(nums1,k1))
print(top_most_freq(nums2,k2))
```

```python
def top_most_freq(nums, k):
    freq = {}

    for n in nums:
        freq[n] = freq.get(n, 0) + 1

    # Buckets where index = frequency
    buckets = [[] for _ in range(len(nums) + 1)]

    for num, count in freq.items():
        buckets[count].append(num)

    res = []
    for i in range(len(buckets) - 1, 0, -1):
        for num in buckets[i]:
            res.append(num)
            if len(res) == k:
                return res
```
