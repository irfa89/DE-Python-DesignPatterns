**1. Opposite Ends**

Problem Statement
Given a sorted array of integers nums and a target sum target, find all pairs of numbers whose sum equals target. Return the pairs as a list of tuples. Use the two pointers approach starting from opposite ends.

Sample Test Cases

Input: nums = [1, 2, 3, 4, 6], target = 6
Output: [(2, 4), (1, 5)]  
Input: nums = [2, 3, 5, 7, 8, 11], target = 10
Output: [(2, 8), (3, 7)]

```python
nums1 = [1, 2, 3, 4, 5, 6]
target1 = 6
nums2 = [2, 3, 5, 7, 8, 11]
target2 = 10

def has_pair(nums,target):
    l, r, res = 0, len(nums) - 1, []

    while l < r:
        s = nums[l] + nums[r]
        if s == target:
            res.append((nums[l],nums[r]))
        if s > target:
            r -= 1
        else:
            l += 1
    return res

print(has_pair(nums1,target1))
print(has_pair(nums2,target2))
```

1. Pair With Target Sum (Sorted Array)

Problem Statement
Given a sorted integer array nums and an integer target, determine whether there exists a pair of elements whose sum equals the target. Return true if such a pair exists, otherwise return false.

Sample Test Cases

Input: nums = [1, 2, 3, 4, 6], target = 6
Output: true

Input: nums = [2, 3, 5, 9, 11], target = 10
Output: false

```python
nums1 = [1, 2, 3, 4, 6]
target1 = 6
nums2 = [2, 3, 5, 9, 11]
target2 = 10

def has_pair(nums,target):
    l, r = 0, len(nums) - 1

    while l < r:
        s = nums[l] + nums[r]
        if s == target:
            return True
        if s > target:
            r -= 1
        else:
            l += 1
    return False

print(has_pair(nums1,target1))
print(has_pair(nums2,target2))
```

2. Reverse String In-Place

Problem Statement
Given a character array s, reverse the array in place using constant extra space.

Sample Test Cases

Input: s = ["h", "e", "l", "l", "o"]
Output: ["o", "l", "l", "e", "h"]

Input: s = ["a"]
Output: ["a"]

```python
s1 = ["h", "e", "l", "l", "o"]
s2 = ["a"]

def reverse_string(s):
    l , r = 0, len(s) - 1
    while l < r:
        s[l] , s[r] = s[r] , s[l]
        l += 1
        r -= 1
    return s

print(reverse_string(s1))
print(reverse_string(s2))
```

3. Valid Palindrome (Ignoring Non-Alphanumeric)

Problem Statement
Given a string s, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

Sample Test Cases

Input: s = "A man, a plan, a canal: Panama"
Output: true

Input: s = "race a car"
Output: false

```python
s1 = "A man, a plan, a canal: Panama"
s2 = "race a car"

def valid_palindrome(s):
    l, r  = 0, len(s) - 1
    while l < r:
        if not s[l].isalnum():
            l += 1
            continue
        if not s[r].isalnum():
            r -= 1
            continue
        if s[l].lower() != s[r].lower() :
            return False
        else:
            l += 1
            r -= 1
    return True

print(valid_palindrome(s1))
print(valid_palindrome(s2))
```

1. Remove Duplicates from Sorted Array

Problem Statement
Given a sorted integer array nums, remove the duplicates in place such that each unique element appears only once. Return the number of unique elements.

Sample Test Cases

Input: nums = [1, 1, 2]
Output: 2

Input: nums = [0, 0, 1, 1, 1, 2, 2, 3]
Output: 4

```python
nums1 = [1, 1, 2]
nums2 = [0, 0, 1, 1, 1, 2, 2, 3]

def remove_duplicates(nums):
    if not nums:
        return 0
    slow = 0
    for fast in range(len(nums)):
        if nums[slow] != nums[fast]:
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1

print(remove_duplicates(nums1))
print(remove_duplicates(nums2))
```

2. Move Zeroes to End

Problem Statement
Given an integer array nums, move all 0s to the end of the array while maintaining the relative order of the non-zero elements.

Sample Test Cases

Input: nums = [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]

Input: nums = [0, 0, 0]
Output: [0, 0, 0]

```python
nums1 = [0, 1, 0, 3, 12]
nums2 = [0, 0, 0]

def move_zeroes(nums):
    slow = 0
    for fast in range(len(nums)):
        if nums[fast] != 0:
            nums[slow],nums[fast] = nums[fast],nums[slow]
            slow += 1
    return nums

print(move_zeroes(nums1))
print(move_zeroes(nums2))
```

1. Merge Two Sorted Arrays

Problem Statement
Given two sorted integer arrays nums1 and nums2, merge them into a single sorted array and return the result.

Sample Test Cases

Input: nums1 = [1, 3, 5], nums2 = [2, 4, 6]
Output: [1, 2, 3, 4, 5, 6]

Input: nums1 = [], nums2 = [1, 2]
Output: [1, 2]

```python
nums11 = [1, 3, 5]
nums12 = [2, 4, 6]
nums21 = []
nums22 = [1, 2]

def merge(a,b):
    i = j = 0
    res = []
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1
    return res + a[i:] + b[j:]

print(merge(nums11,nums12))
print(merge(nums21,nums22))
```
