**1. Fixed Window**

1. Maximum Sum Subarray of Size K (Classic)

Problem Statement
Given an integer array nums and an integer k, find the maximum sum of any contiguous subarray of size k.

Sample Test Cases

Input: nums = [2, 1, 5, 1, 3, 2], k = 3
Output: 9
Explanation: Subarray [5,1,3]

Input: nums = [2, 3, 4, 1, 5], k = 2
Output: 7
Explanation: Subarray [3,4]

```python
nums1 = [2, 1, 5, 1, 3, 2]
nums2 = [2, 3, 4, 1, 5]
k1 = 3
k2 = 2

def max_sum_subarr(nums,k):
  l, window_sum , max_sum = 0, 0 , float('-inf')
  for r in range(len(nums)):
    window_sum += nums[r]
    if r - l + 1 == k:
      max_sum = max(window_sum,max_sum)
      window_sum -= nums[l]
      l += 1
  return max_sum

print(max_sum_subarr(nums1,k1))
print(max_sum_subarr(nums2,k2))
```

7. Maximum Sum of Fixed Window in Circular Array

Problem Statement
Given a circular array of integers and an integer k, find the maximum sum of any contiguous subarray of size k.

Sample Test Cases

Input: nums = [5, -2, 3, 4], k = 3
Output: 12
Explanation: Circular subarray [3,4,5]

Input: nums = [8, -1, 3, 4], k = 2
Output: 12

```python
nums1 = [5, -2, 3, 4]
k1 = 3
nums2 = [8, -1, 3, 4]
k2 = 2

def max_sum_circular(nums,k):
    n = len(nums)
    window_sum = sum(nums[:k])
    max_sum = window_sum
    for r in range(1,n):
        window_sum -= nums[r - 1]
        window_sum += nums[(r + k - 1) % n]  # Modulo implements circular window
        max_sum = max(max_sum,window_sum)
    return max_sum

print(max_sum_circular(nums1,k1))
print(max_sum_circular(nums2,k2)) 
```

**2. Variable Window (At Most)**

1. Longest Substring with At Most K Distinct Characters (Classic)

Problem Statement
Given a string s and an integer k, find the length of the longest substring that contains at most k distinct characters.

Sample Test Cases

Input: s = "eceba", k = 2
Output: 3

Input: s = "aa", k = 1
Output: 2

```python
s1 = "eceba"
k1 = 2

s2 = "aa"
k2 = 1

def longest_substr_k_distinct(s,k):
    if k == 0:
        return 0
    freq = {}
    l = 0
    max_len = 0

    for r in range(len(s)):
        freq[s[r]] = freq.get(s[r],0) + 1
        while len(freq) > k:
            freq[s[l]] -= 1
            if freq[s[l]] == 0:
                del freq[s[l]]
            l += 1
            
        max_len = max(max_len,r - l + 1)
    return max_len

print(longest_substr_k_distinct(s1,k1))
print(longest_substr_k_distinct(s2,k2))
```

2. Longest Substring with At Most One Distinct Character

Problem Statement
Given a string s, return the length of the longest substring that contains at most one distinct character.

Sample Test Cases

Input: s = "abbbcc"
Output: 3

Input: s = "aaaa"
Output: 4

```python
s1 = "abbbcc"
s2 = "aaaa"

def longest_substr_one_distinct(s):
    l,freq, max_len = 0,{},0

    if not s:
        return 0

    for r in range(len(s)):
        freq[s[r]] = freq.get(s[r],0) + 1

        while len(freq) > 1: # # shrink window if distinct characters > 1
            freq[s[l]] -= 1
            if freq[s[l]] == 0:
                del freq[s[l]]
            l += 1

        max_len = max(max_len,r - l + 1)
    return max_len

print(longest_substr_one_distinct(s1))
print(longest_substr_one_distinct(s2))
```

3. Longest Substring with At Most K Distinct Characters (All Unique Case)

Problem Statement
Given a string where all characters may be different, find the length of the longest substring with at most k distinct characters.

Sample Test Cases

Input: s = "abcdef", k = 3
Output: 3

Input: s = "abcd", k = 1
Output: 1

```python
s1 = "abcdef"
k1 = 3
s2 = "abcd"
k2 = 1

def longest_substr_k_distinct(s,k):
    l,max_len,freq = 0,0,{}

    if k==0 or not s:
        return 0

    for r in range(len(s)):
        freq[s[r]] = freq.get(s[r],0) + 1
        if len(freq) > k:
            freq[s[l]] -= 1
            if freq[s[l]] == 0:
                del freq[s[l]]
            l += 1
        max_len = max(max_len,r - l + 1)
    return max_len

print(longest_substr_k_distinct(s1,k1))
print(longest_substr_k_distinct(s2,k2))
```
