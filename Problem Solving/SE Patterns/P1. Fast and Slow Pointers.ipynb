**1. Cycle Detection**

**1. Linked List Cycle Detection**

Problem Statement
Given the head of a singly linked list, determine if the linked list contains a cycle.
A cycle exists if a node can be reached again by continuously following the next pointer.

Sample Test Cases
Input: head = [3, 2, 0, -4], pos = 1
Output: true

Input: head = [1, 2], pos = -1
Output: false

**2. Length of Cycle in a Linked List**

Problem Statement
Given the head of a singly linked list that may contain a cycle, return the length of the cycle.
If there is no cycle, return 0.

Sample Test Cases
Input: head = [1, 2, 3, 4, 5], pos = 2
Output: 3

Input: head = [1, 2, 3], pos = -1
Output: 0

**3. Start of Cycle in a Linked List**

Problem Statement
Given the head of a singly linked list, return the node where the cycle begins.
If there is no cycle, return null.

Sample Test Cases
Input: head = [1, 2, 3, 4, 5], pos = 1
Output: Node with value 2

Input: head = [1, 2, 3], pos = -1
Output: null

**4. Happy Number**

Problem Statement
A number is called happy if, by repeatedly replacing it with the sum of the squares of its digits, it eventually becomes 1.
If it enters a cycle that does not include 1, it is not a happy number.
Given an integer n, return true if it is a happy number, otherwise return false.

Sample Test Cases
Input: n = 19
Output: true

Input: n = 2
Output: false

**5. Circular Array Loop**

Problem Statement
Given a circular array of non-zero integers where each element represents the number of steps to move forward or backward, determine if there exists a cycle.
A valid cycle must have more than one element and must move in a single direction.

Sample Test Cases
Input: nums = [2, -1, 1, 2, 2]
Output: true

Input: nums = [-1, 2]
Output: false
