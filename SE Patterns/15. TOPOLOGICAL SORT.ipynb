**‚≠ê 4. Canonical Template (üî• MEMORIZE THIS üî•)**

```python
from collections import deque, defaultdict

def topological_sort(graph):
    indegree = {u:0 for u in graph}
    for u in graph:
        for v in graph[u]:
            indegree[v] += 1

    q = deque([u for u in graph if indegree[u]==0])
    order = []

    while q:
        node = q.popleft()
        order.append(node)
        for neighbor in graph[node]:
            indegree[neighbor] -= 1
            if indegree[neighbor]==0:
                q.append(neighbor)

    return order if len(order)==len(graph) else []  # cycle check
```

1. DFS-based Topo Sort

Variation: DFS

Change: recursion, post-order append, reverse at end

```python
def topo_dfs(graph):
    visited = set()
    res = []
    def dfs(node):
        visited.add(node)
        for nei in graph[node]:
            if nei not in visited:
                dfs(nei)
        res.append(node)  # post-order append
    for u in graph:
        if u not in visited:
            dfs(u)
    return res[::-1]
```

3. Cycle Detection

Variation: BFS + cycle validation

```python
def topo_cycle_check(graph):
    indegree = {u:0 for u in graph}
    for u in graph:
        for v in graph[u]:
            indegree[v] += 1
    q = deque([u for u in graph if indegree[u]==0])
    order = []
    while q:
        node = q.popleft()
        order.append(node)
        for n in graph[node]:
            indegree[n]-=1
            if indegree[n]==0:
                q.append(n)
    return order if len(order)==len(graph) else []  # empty if cycle
```
