2. Permutations

Q: Generate all permutations of [1,2,3].

```python
# Variation: mark used choices
def backtrack(path):
    if len(path)==len(nums):
        output.append(path[:])
        return
    for i in range(len(nums)):
        if used[i]:
            continue
        used[i]=True
        path.append(nums[i])
        backtrack(path)
        path.pop()
        used[i]=False
nums=[1,2,3]; used=[False]*len(nums); output=[]
backtrack([])
print(output)
```

3. Combination Sum / Target Sum

Q: Combinations from [2,3,6,7] sum to 7.

```python
def backtrack(path, start, total):
    if total==target:
        output.append(path[:]); return
    if total>target: return
    for i in range(start,len(nums)):
        path.append(nums[i])
        backtrack(path,i,total+nums[i])
        path.pop()
nums=[2,3,6,7]; target=7; output=[]
backtrack([],0,0)
print(output)
```

4. Palindrome Partitioning

Q: Partition "aab" into all palindromes.
