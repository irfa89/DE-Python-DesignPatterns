**⭐ 1. What This Pattern Solves (INTERVIEW SIGNALS)**

Problem involves strings, especially a large set of words or prefixes.
Lookups like search, insert, startsWith, autocomplete.
Input constraints: lots of repeated prefixes; naive O(n*m) search is too slow.
Brute-force: scanning all strings sequentially → inefficient for prefix-heavy datasets.
Typical indicators: “implement a dictionary”, “check word existence”, “find all words with given prefix”.

**⭐ 3. Core Idea (MAX 2 LINES)**

Each node stores children for the next character; traverse down the tree to insert/search.
Prefixes are shared, reducing redundant storage and enabling fast lookups.


**⭐ 4. Canonical Template (🔥 MEMORIZE THIS 🔥)**

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end

    def startsWith(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True
```

**⭐ 5. Pattern Variations (COMPLETE LIST — MUST COVER ALL)**

Standard Trie (Insert/Search)
Insert/search exact words.
Template: as above.
Prefix Search / startsWith
Search only for prefixes.
Change: return True after traversing prefix, no need to check is_end.
Autocomplete / Collect All Words With Prefix
Return all words with a given prefix.
Change: DFS/BFS from prefix node to collect words.
Trie with Counts / Frequency Map
Track frequency of words/prefixes.
Change: add count attribute in nodes, increment during insert.
Deletion from Trie
Remove a word safely without breaking other prefixes.
Change: recursive delete with pruning empty nodes.

**⭐ 6. Worked Example (Canonical Template)**

Input: ["apple", "app", "apt"], search "app", startsWith "ap"

Step-by-step:

Insert "apple": a → p → p → l → e, mark end.
Insert "app": reuse a → p → p, mark end.
Insert "apt": reuse a → p, add t, mark end.
search("app") → traverse a → p → p → is_end=True → return True.
startsWith("ap") → traverse a → p → exists → return True.

```python
search("app") = True
startsWith("ap") = True
search("bat") = False
```

**⭐ 7. Variation-Based Solved Coding Questions (MANDATORY)**

1. Standard Trie (Insert/Search)

```python
# Variation: exact search
trie = Trie()
for word in ["cat","car","dog"]: trie.insert(word)
print(trie.search("car"))  # True
print(trie.search("ca"))   # False
```

2. Prefix Search / startsWith

```python
# Variation: prefix check
print(trie.startsWith("ca"))  # True
print(trie.startsWith("do"))  # True
print(trie.startsWith("da"))  # False
```

3. Autocomplete / Collect All Words With Prefix

```python
# Variation: DFS from prefix
def autocomplete(trie, prefix):
    node = trie.root
    for char in prefix:
        if char not in node.children: return []
        node = node.children[char]
    results = []
    def dfs(n, path):
        if n.is_end: results.append(prefix+path)
        for c, child in n.children.items(): dfs(child, path+c)
    dfs(node, "")
    return results

autocomplete(trie, "ca")  # ["cat","car"]
```

4. Trie with Counts / Frequency Map

```python
# Variation: count occurrences
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False
        self.count = 0

trie = Trie()
words = ["apple","app","apple"]
for w in words:
    node = trie.root
    for c in w:
        if c not in node.children: node.children[c]=TrieNode()
        node = node.children[c]
    node.is_end=True
    node.count+=1
# node.count now shows frequency
```

5. Deletion from Trie

```python
# Variation: safe delete
def delete(node, word, index=0):
    if index == len(word):
        if node.is_end: node.is_end=False
        return len(node.children)==0
    char = word[index]
    if char not in node.children: return False
    should_delete = delete(node.children[char], word, index+1)
    if should_delete:
        del node.children[char]
        return len(node.children)==0 and not node.is_end
    return False

delete(trie.root, "app")
```

**⭐ 8. Time & Space Complexity (INTERVIEW READY)**

Insert/Search: O(L) where L = word length.
startsWith: O(P) where P = prefix length.
Autocomplete DFS: O(P + N*L') worst-case, N = words under prefix, L' = avg length.
Space: O(N*L) for N words of avg length L, optimized by shared prefixes.

**⭐ 9. Common Failure Modes (WHY CANDIDATES FAIL)**

❌ Forgetting to mark is_end, causing search to fail.
❌ Infinite recursion in DFS for autocomplete.
❌ Deleting nodes without checking other branches → broken words.
❌ Using mutable default dict incorrectly, sharing state between nodes.

✔ Mental corrections: always check is_end, traverse carefully, prune only if safe, use per-node children dict.
