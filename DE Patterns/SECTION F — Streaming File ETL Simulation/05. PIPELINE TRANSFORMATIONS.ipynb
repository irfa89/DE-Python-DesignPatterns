**⭐ 3. Core Idea**

Apply a sequence of deterministic transformations on each record to prepare it for downstream aggregation or storage.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
# Input: iterable of dicts or objects
data = [
    {"name": "alice", "age": "25", "salary": 50000, "active": True},
    {"name": "bob", "age": "30", "salary": 60000, "active": False},
]

# Transformation functions
def filter_active(record):
    return record["active"]

def normalize_name(record):
    record["name"] = record["name"].upper()
    return record

def cast_age(record):
    record["age"] = int(record["age"])
    return record

# Pipeline
pipeline = [filter_active, normalize_name, cast_age]

# Apply transformations
transformed = [
    func(record) for record in data if all(func(record) for func in pipeline)
]
```

**⭐ 9. Common Pitfalls & Mistakes**

❌ Mutating shared objects across transformations (causes side effects)
❌ Forgetting to filter before expensive transformations (wastes CPU)
❌ Applying transformations in the wrong order (e.g., casting before filtering)
✔ Use pure functions where possible
✔ Apply filters first to reduce downstream load
✔ Chain transformations clearly for readability and maintainability
