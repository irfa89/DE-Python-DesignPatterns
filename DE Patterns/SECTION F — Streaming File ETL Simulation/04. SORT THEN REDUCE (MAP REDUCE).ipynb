**⭐ 1. What This Pattern Solves**

Efficiently aggregates large datasets where items need grouping by key.
Useful for distributed ETL pipelines where in-memory grouping is expensive.
Handles operations like sum, count, average, min/max after sorting/grouping.
Common in analytics pipelines for logs, metrics, and transactional data.

**⭐ 2. SQL Equivalent**

```python
%sql
SELECT key, SUM(value) AS total_value
FROM table
GROUP BY key
ORDER BY key;
```

**⭐ 3. Core Idea**

Sort data by key, then reduce consecutive items per key.
Works because all items with the same key become contiguous.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
from itertools import groupby
from operator import itemgetter

# data = [(key1, value1), (key2, value2), ...]
data.sort(key=itemgetter(0))  # Sort by key

result = []
for key, group in groupby(data, key=itemgetter(0)):
    reduced_value = sum(v for _, v in group)  # example reduction
    result.append((key, reduced_value))
```

**⭐ 5. Detailed Example**

```python
data = [("a", 2), ("b", 3), ("a", 5), ("b", 7), ("c", 1)]

from itertools import groupby
from operator import itemgetter

data.sort(key=itemgetter(0))
result = [(k, sum(v for _, v in g)) for k, g in groupby(data, key=itemgetter(0))]
print(result)

[('a', 7), ('b', 10), ('c', 1)]
```

**⭐ 7. Full Data Engineering Scenario**

Problem Statement:
You have a log of page visits: (user_id, page_id, duration_seconds). Compute total time spent per page.

Expected Output:
[(page_id1, total_time1), (page_id2, total_time2), ...]

Skeleton Solution:

```python
# logs = [(user_id, page_id, duration), ...]
page_durations = [(page_id, duration) for _, page_id, duration in logs]
page_durations.sort(key=itemgetter(0))

result = [(pid, sum(d for _, d in g)) for pid, g in groupby(page_durations, key=itemgetter(0))]
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(N log N) due to sort, then O(N) for reduce → total O(N log N)
Space Complexity: O(N) for storing intermediate sorted data and result

**⭐ 9. Common Pitfalls & Mistakes**

❌ Forgetting to sort before groupby → wrong grouping
❌ Using groupby on unsorted data → partial or incorrect aggregation
✔ Always sort by the same key used in groupby
✔ Use generator expressions for memory-efficient reductions
✔ Avoid excessive intermediate list creations for large datasets
