**⭐ 1. What This Pattern Solves**

Efficiently process very large files (GBs/TBs) without loading entire content into memory.
Enable ETL pipelines to handle line-by-line or chunked transformations.
Support continuous ingestion for logs, CSVs, JSON Lines, or other append-only files.
Reduce memory footprint while allowing aggregation, filtering, or transformation on-the-fly.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Conceptually, streaming a file line-by-line is like:
SELECT *
FROM large_table
WHERE condition
-- applied row-by-row instead of loading all rows at once.
```

**⭐ 3. Core Idea**

Read and process data incrementally instead of all at once; keep memory usage constant.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
# Stream a text file line by line
with open("path/to/file.txt", "r") as f:
    for line in f:
        process(line)  # transform, filter, or aggregate

# Stream CSV in chunks
import pandas as pd
for chunk in pd.read_csv("path/to/file.csv", chunksize=10000):
    process(chunk)
```

**⭐ 7. Full Data Engineering Scenario**

Problem Statement: You receive a daily 50GB transaction log in CSV. You need to compute total sales per product without crashing memory.

Expected Output: Dictionary {product_id: total_sales}

Skeleton Solution:

```python
import pandas as pd
from collections import defaultdict

sales = defaultdict(float)

for chunk in pd.read_csv("daily_transactions.csv", chunksize=500000):
    for _, row in chunk.iterrows():
        sales[row['product_id']] += row['amount']
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n) – each row is read and processed once.
Space Complexity: O(k) – memory depends on aggregation storage, not file size (k = number of unique keys).

**⭐ 9. Common Pitfalls & Mistakes**

❌ Loading entire file into memory → crashes on large files
❌ Forgetting to use chunksize with pandas → memory spikes
❌ Inefficient per-line operations without vectorization → slows down ETL
✔ Use generators or chunked reading
✔ Aggregate incrementally
✔ Only store necessary data in memory
