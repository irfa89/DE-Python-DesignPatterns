**⭐ 1. What This Pattern Solves**

Handles cumulative calculations over a stream or sequence of events.
Tracks running totals, averages, counts, or sums in ETL pipelines.
Useful for analytics where intermediate state matters (e.g., rolling metrics, session stats).
Efficiently avoids recomputing from scratch on every new record.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Running total per user
SELECT
    user_id,
    event_time,
    SUM(value) OVER (PARTITION BY user_id ORDER BY event_time ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM events;
```

**⭐ 3. Core Idea**

Keep an evolving “state” object updated as new records arrive, avoiding full recomputation.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
from collections import defaultdict

# Initialize state
state = defaultdict(int)

for record in stream:  # stream can be list, generator, or file iterator
    key = record['key']
    value = record['value']
    
    # Update state
    state[key] += value
    
    # Optional: emit intermediate result
    print(key, state[key])
```

**⭐ 5. Detailed Example**

```python
stream = [
    {'key': 'A', 'value': 10},
    {'key': 'B', 'value': 5},
    {'key': 'A', 'value': 3},
    {'key': 'B', 'value': 7},
]

from collections import defaultdict

state = defaultdict(int)

for record in stream:
    k = record['key']
    v = record['value']
    state[k] += v
    print(f"{k}: {state[k]}")

A: 10
B: 5
A: 13
B: 12
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n) — each record updates state once.
Space Complexity: O(k) — where k = number of unique keys in the stream.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Recomputing totals from scratch for every new record.
❌ Using a regular dict without defaultdict and manually checking keys.
❌ Forgetting to update state for each record in loops/generators.
✔ Always update cumulative state incrementally.
✔ Use defaultdict for concise initialization.
✔ Emit intermediate or final state as needed, avoiding unnecessary recomputation.
