**⭐ 1. What This Pattern Solves**

Often in ETL pipelines, we get nested JSON or semi-structured data containing arrays/lists of objects.
We need to flatten arrays to normalize data for analytics or storage in relational tables.
Useful for exploding logs, events, transaction details into rows.
Handles nested arrays, filtering, and aggregation post-flatten.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Flattening an array column into rows
SELECT id, array_element
FROM my_table
CROSS JOIN UNNEST(array_column) AS t(array_element);
```

**⭐ 3. Core Idea**

Convert array/list elements into individual rows for processing.
Treat each element as a separate record for transformations, aggregations, or joins.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
# Python: Extract / Flatten Array Fields
from itertools import chain

# input_data = [{'id': 1, 'values': [10, 20]}, {'id': 2, 'values': [30]}]
flattened = [{'id': d['id'], 'value': v} for d in input_data for v in d['values']]

# nested arrays deeper than 1 level
# flattened_deep = [{'id': d['id'], 'value': v} for d in input_data for sublist in d['values'] for v in sublist]
```

**⭐ 5. Detailed Example**

```python
input_data = [
    {'user_id': 101, 'purchases': ['apple', 'banana']},
    {'user_id': 102, 'purchases': ['orange']}
]

flattened = [{'user_id': d['user_id'], 'purchase': p} for d in input_data for p in d['purchases']]

[
    {'user_id': 101, 'purchase': 'apple'},
    {'user_id': 101, 'purchase': 'banana'},
    {'user_id': 102, 'purchase': 'orange'}
]
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n * m) → n = number of parent objects, m = avg array size.
Space Complexity: O(n * m) → output list has one row per array element.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Forgetting to handle empty arrays, resulting in missing rows.
❌ Using nested loops inefficiently for deeply nested arrays.
✔ Use list comprehension or itertools.chain.from_iterable for flattening.
✔ Validate array fields exist to avoid KeyError.
✔ Avoid repeated deep copies; work with references when possible.
