**⭐ 1. What This Pattern Solves**

Safely accessing deeply nested dictionaries or JSON without raising KeyError.
Avoids multiple if/try-except checks in ETL pipelines when reading optional fields.
Useful in transformation layers where some records may have missing nested data.
Enables robust analytics and aggregation without breaking on missing keys.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Using COALESCE with JSON fields in SQL
SELECT COALESCE(data->'user'->>'name', 'Unknown') AS user_name
FROM events;
```

COALESCE provides default if path is missing.

JSON operators mimic nested dictionary access.

**⭐ 3. Core Idea**

Chain .get() calls with defaults to traverse nested structures safely.
Return a fallback if any key in the chain is missing.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
# TEMPLATE
def safe_get_chain(d, keys, default=None):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key, default)
        else:
            return default
    return d

# Example Usage
value = safe_get_chain(record, ['user', 'profile', 'age'], default=0)
```

**⭐ 5. Detailed Example**

```python
record = {
    'user': {
        'profile': {
            'name': 'Alice'
        }
    }
}

age = safe_get_chain(record, ['user', 'profile', 'age'], default=30)
name = safe_get_chain(record, ['user', 'profile', 'name'], default='Unknown')

age   # 30
name  # 'Alice'
```

**⭐ 7. Full Data Engineering Scenario**

Problem Statement:
You ingest JSON logs where user.device.os.version may be missing. You need to count all users by OS version without crashing on missing fields.

Expected Output:

```python
{'Android 12': 120, 'iOS 16': 95, 'Unknown': 30}

from collections import Counter

os_counter = Counter()
for log in logs:
    version = safe_get_chain(log, ['user', 'device', 'os', 'version'], default='Unknown')
    os_counter[version] += 1
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n * k), n = number of records, k = depth of keys.
Space Complexity: O(1) extra space per lookup (aside from storing results).

**⭐ 9. Common Pitfalls & Mistakes**

❌ Using multiple nested try-except instead of .get() → verbose and error-prone.
❌ Assuming all values are dictionaries; can raise AttributeError if a key returns non-dict.
✔ Always check type before .get() to avoid breaking on non-dict values.
✔ Provide sensible default for analytics aggregation to prevent missing values.
