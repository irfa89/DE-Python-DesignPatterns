**⭐ 1. What This Pattern Solves**

Extract relevant entries from complex nested JSON structures.
Apply conditions or filters to objects inside arrays or nested dictionaries.
Useful in ETL pipelines when only a subset of deeply nested data is required.
Reduces payload size before transformations or aggregations.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Imagine JSON flattened into a table with nested array values
SELECT *
FROM orders o
JOIN LATERAL UNNEST(o.items) AS item
WHERE item.price > 100;
```

**⭐ 3. Core Idea**

Traverse nested structures, evaluate a condition, and keep only matching elements. Works because you process each node independently and reconstruct the structure.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
def filter_nested_json(data, condition):
    if isinstance(data, list):
        return [filter_nested_json(x, condition) for x in data if condition(x)]
    elif isinstance(data, dict):
        return {k: filter_nested_json(v, condition) for k, v in data.items() if filter_nested_json(v, condition) != {}}
    else:
        return data
```

**⭐ 5. Detailed Example**

```python
orders = [
    {"id": 1, "items": [{"name": "apple", "qty": 5}, {"name": "banana", "qty": 2}]},
    {"id": 2, "items": [{"name": "apple", "qty": 1}, {"name": "orange", "qty": 10}]}
]

def condition(item):
    return isinstance(item, dict) and item.get("qty", 0) >= 5

filtered_orders = filter_nested_json(orders, condition)
print(filtered_orders)

[
    {"id": 1, "items": [{"name": "apple", "qty": 5}]},
    {"id": 2, "items": [{"name": "orange", "qty": 10}]}
]
```

**⭐ 9. Common Pitfalls & Mistakes**

❌ Modifying the list/dict in place while iterating → skips elements or crashes.
❌ Not handling mixed types (dict vs list vs primitive) → TypeError.
❌ Forgetting to remove empty containers after filtering → leftover empty dicts/lists.

✔ Always build a new structure or use safe comprehensions.
✔ Explicitly check type before recursion.
✔ Ensure condition works for nested objects only.
