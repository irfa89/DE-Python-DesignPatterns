**⭐ 1. What This Pattern Solves**

Navigating nested dictionaries or JSON-like structures in pipelines.
Extracting all keys (or key paths) for transformation, flattening, or validation.
Useful in ETL for schema discovery, auditing, or recursive transformations.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Conceptually, there is no direct SQL recursion for nested dicts
-- But similar to unnesting JSON in SQL:
SELECT json_object_keys(data) AS key
FROM nested_json_table;
```

**⭐ 3. Core Idea**

Recursively traverse nested dicts.
Collect keys (or key paths) in a DFS manner.
Works because dictionaries are naturally tree-like structures.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
def dfs_keys(d, parent_key=""):
    keys = []
    for k, v in d.items():
        full_key = f"{parent_key}.{k}" if parent_key else k
        keys.append(full_key)
        if isinstance(v, dict):
            keys.extend(dfs_keys(v, full_key))
    return keys
```

**⭐ 5. Detailed Example**

```python
data = {
    "user": {
        "id": 1,
        "info": {
            "name": "Alice",
            "email": "alice@example.com"
        }
    },
    "meta": {
        "timestamp": "2025-12-16"
    }
}

all_keys = dfs_keys(data)
print(all_keys)

['user', 'user.id', 'user.info', 'user.info.name', 'user.info.email', 'meta', 'meta.timestamp']
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(N) — each key/value in the nested dict is visited once.
Space Complexity: O(N) — list of all keys plus recursion stack.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Forgetting to handle nested dicts → only top-level keys returned
❌ Using mutable default args (keys=[]) → shared state bugs
✔ Correct: pass accumulated keys as local list and extend recursively
✔ Use isinstance(v, dict) to safely check nesting
✔ Optionally handle lists of dicts if JSON contains arrays of objects
