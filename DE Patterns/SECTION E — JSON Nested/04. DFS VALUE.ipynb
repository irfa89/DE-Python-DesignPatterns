**⭐ 1. What This Pattern Solves**

Extract all values from a nested dictionary or JSON-like structure.
Useful in ETL pipelines when flattening hierarchical data.
Helps in analytics pipelines where only leaf values matter.
Supports mixed nesting: dicts, lists, or combinations.

**⭐ 2. SQL Equivalent**

No direct SQL equivalent, but conceptually similar to:

**⭐ 3. Core Idea**

Recursively traverse nested structures.
Yield or collect values only (skip keys).
Works for arbitrarily deep dictionaries and lists.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
def dfs_values(node):
    if isinstance(node, dict):
        for v in node.values():
            yield from dfs_values(v)
    elif isinstance(node, list):
        for item in node:
            yield from dfs_values(item)
    else:
        yield node

# Usage:
# values = list(dfs_values(nested_dict))
```

**⭐ 5. Detailed Example**

```python
data = {
    'a': 1,
    'b': {'c': 2, 'd': [3, {'e': 4}]},
    'f': [5, 6]
}

list(dfs_values(data))

[1, 2, 3, 4, 5, 6]
```

Step-by-step:

a -> 1 → collect 1
b.c -> 2 → collect 2
b.d[0] -> 3 → collect 3
b.d[1].e -> 4 → collect 4
f[0] -> 5, f[1] -> 6 → collect 5,6

**⭐ 8. Time & Space Complexity**

Time Complexity: O(N) — every element is visited once.
Space Complexity: O(D) recursion stack, where D = max depth; output list = O(M), M = number of leaf nodes.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Forgetting to handle lists inside dicts → misses values in arrays.
❌ Using append instead of yield from → nested lists instead of flat.
❌ Exceeding recursion depth for very deep structures.

✔ Correct approach: recursion with yield from, handle dicts & lists, flatten on collection.
