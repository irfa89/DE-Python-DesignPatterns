**⭐ 1. What This Pattern Solves**

Combines multiple nested dictionaries or JSON objects into a single unified structure.
Resolves key conflicts by merging sub-dictionaries instead of overwriting.
Useful in:
ETL pipelines ingesting hierarchical JSON logs or configs.
Aggregating user profiles from multiple sources.
Merging metrics or event payloads before analytics.

**⭐ 3. Core Idea**

Recursively combine keys; if both are dicts, merge deeper; else, overwrite or aggregate values. Works because dictionaries map hierarchically.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
def merge_nested(a: dict, b: dict) -> dict:
    result = a.copy()
    for k, v in b.items():
        if k in result and isinstance(result[k], dict) and isinstance(v, dict):
            result[k] = merge_nested(result[k], v)
        else:
            result[k] = v
    return result
```

**⭐ 5. Detailed Example**

```python
a = {"user": {"id": 1, "name": "Alice"}, "metrics": {"views": 10}}
b = {"user": {"email": "alice@example.com"}, "metrics": {"clicks": 5}}

merged = merge_nested(a, b)
print(merged)

{
    "user": {"id": 1, "name": "Alice", "email": "alice@example.com"},
    "metrics": {"views": 10, "clicks": 5}
}
```

**⭐ 9. Common Pitfalls & Mistakes**

❌ Overwriting nested dictionaries instead of merging.
❌ Ignoring type mismatches (dict vs list vs scalar).
❌ Using shallow copy which mutates input dicts.

✔ Use recursive merge.
✔ Decide on conflict strategy: overwrite, append to list, or aggregate.
✔ Always copy input to avoid side effects.
