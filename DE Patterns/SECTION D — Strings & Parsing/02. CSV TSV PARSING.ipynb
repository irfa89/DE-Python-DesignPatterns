**⭐ 1. What This Pattern Solves**

Handles ingestion of raw CSV or TSV files into Python pipelines.
Converts flat text rows into structured dictionaries or lists for transformations.
Supports ETL workflows: cleaning, type conversion, aggregation, and analytics.
Useful for pipelines where input files are delimited but not preloaded in databases.

```python
%sql
-- Reading CSV/TSV is like creating a temporary table from external data
CREATE TEMP TABLE raw_data AS
SELECT *
FROM external_csv_file;
```

**⭐ 3. Core Idea**

Read lines → split by delimiter → map fields to headers → optional type conversion.
Works because structured parsing allows downstream transformations.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
import csv

# Read CSV/TSV into list of dicts
def parse_delimited_file(file_path: str, delimiter: str = ','):
    with open(file_path, 'r', newline='') as f:
        reader = csv.DictReader(f, delimiter=delimiter)
        return [dict(row) for row in reader]

# Example usage
data = parse_delimited_file('data.csv', delimiter=',')
```

**⭐ 7. Full Data Engineering Scenario**

Problem Statement:
Daily log files are exported as CSV with columns: user_id, event_type, timestamp. The goal is to aggregate total events per user for analytics.

Expected Output:

```python
{'user_1': 10, 'user_2': 7, ...}
```

```python
logs = parse_delimited_file('daily_logs.csv')
event_counts = {}
for row in logs:
    user = row['user_id']
    event_counts[user] = event_counts.get(user, 0) + 1
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n * m) → n rows, m columns per row.
Space Complexity: O(n) → storing n dictionaries in memory.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Not specifying the correct delimiter → misaligned columns.
❌ Forgetting to handle headers → data shifted to wrong keys.
❌ Keeping numbers/booleans as strings → downstream aggregation errors.
✔ Correct approach: use csv.DictReader with proper delimiter and optional type conversion.
✔ For large files, consider generator-based parsing to reduce memory footprint:

```python
def stream_parse(file_path, delimiter=','):
    with open(file_path, 'r', newline='') as f:
        reader = csv.DictReader(f, delimiter=delimiter)
        for row in reader:
            yield row
```
