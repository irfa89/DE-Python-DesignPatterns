Handles messy, inconsistent, or incomplete data in ETL pipelines.

Reduces errors from missing, duplicate, or malformed records.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Remove duplicates
SELECT DISTINCT *
FROM raw_table;

-- Standardize formats
SELECT UPPER(name) AS name_clean,
       TRIM(email) AS email_clean,
       CAST(age AS INT) AS age_clean
FROM raw_table
WHERE age IS NOT NULL;

-- Replace missing values
SELECT COALESCE(salary, 0) AS salary_clean
FROM raw_table;
```

**⭐ 7. Full Data Engineering Scenario**

Problem:
Ingest daily e-commerce logs with inconsistent product IDs, missing user emails, and mixed-case category names.

Expected Output:

Unique records
Standardized product IDs and categories
No null emails

Skeleton Solution:

```python
logs = pd.read_csv("daily_logs.csv")

# Drop duplicates
logs = logs.drop_duplicates()

# Fill missing emails
logs['user_email'] = logs['user_email'].fillna("unknown@example.com")

# Standardize product_id and category
logs['product_id'] = logs['product_id'].str.upper()
logs['category'] = logs['category'].str.lower().str.strip()
```

**⭐ 8. Time & Space Complexity**

Time: O(n) for single-pass operations like fillna, strip, type cast; O(n log n) if deduplication involves sorting.
Space: O(n) to store the cleaned DataFrame in memory.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Using fillna without considering correct imputation strategy
❌ Applying .str operations on non-string columns
❌ Dropping duplicates too early (may remove needed historical records)
✔ Correct: cast types before string operations, carefully handle missing values, and ensure deduplication preserves critical records.
