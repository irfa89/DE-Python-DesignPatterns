**⭐ 5. Detailed Example**

```python
timestamps = ["2025-12-16 12:34:56", "16-Dec-2025 12:34 PM", "12/16/2025 00:34"]

from datetime import datetime
from dateutil import parser

parsed = [parser.parse(ts) for ts in timestamps]
print(parsed)

[datetime.datetime(2025, 12, 16, 12, 34, 56),
 datetime.datetime(2025, 12, 16, 12, 34),
 datetime.datetime(2025, 12, 16, 0, 34)]
```

**⭐ 7. Full Data Engineering Scenario**

Problem: You receive daily logs from multiple servers in formats like "2025-12-16T12:34:56Z", "16/12/2025 12:34 PM". You need to:

Parse timestamps
Convert all to UTC
Aggregate number of events per hour

```python
Hour        Event_Count
2025-12-16 12:00      245
2025-12-16 13:00      312

from dateutil import parser
import pytz
from collections import Counter

events = ["2025-12-16T12:34:56Z", "16/12/2025 12:34 PM", ...]
timestamps_utc = [parser.parse(ts).astimezone(pytz.UTC) for ts in events]

hourly_counts = Counter(ts.replace(minute=0, second=0, microsecond=0) for ts in timestamps_utc)
```

**⭐ 8. Time & Space Complexity**

Time Complexity: O(n) — each timestamp is parsed once.
Space Complexity: O(n) — storing datetime objects for n timestamps.

**⭐ 9. Common Pitfalls & Mistakes**

❌ Ignoring timezone differences → leads to incorrect aggregations.
❌ Using strptime with a fixed format when data has multiple formats → parsing errors.
❌ Forgetting to handle missing or malformed timestamps → pipeline crashes.
✔ Use dateutil.parser.parse for flexible formats.
✔ Convert all timestamps to a common timezone before aggregation.
✔ Validate timestamp formats and handle exceptions gracefully.
