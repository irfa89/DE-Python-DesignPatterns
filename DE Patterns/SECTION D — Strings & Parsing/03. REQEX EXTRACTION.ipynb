**⭐ 1. What This Pattern Solves**

Extract structured information from semi-structured text (logs, CSVs, JSON strings).
Parse timestamps, IDs, emails, URLs, or key-value data embedded in text fields.
Enable downstream ETL transformations and analytics by converting free-form text into structured columns.

**⭐ 2. SQL Equivalent**

```python
%sql
-- Extract using REGEXP in SQL
SELECT
    REGEXP_SUBSTR(log_message, '[0-9]{4}-[0-9]{2}-[0-9]{2}') AS date,
    REGEXP_SUBSTR(log_message, 'user_id=[0-9]+') AS user_id
FROM logs;
```

**⭐ 3. Core Idea**

Define a regex pattern to capture the desired parts of text; apply it to extract structured data efficiently. Works because regex handles variable and complex text patterns.

**⭐ 4. Template Code (MEMORIZE THIS)**

```python
import re

text = "user_id=1234, date=2025-12-16, action=login"

# Compile pattern (optional, improves performance if reused)
pattern = re.compile(r"user_id=(\d+), date=([\d-]+), action=(\w+)")

# Extract data
match = pattern.search(text)
if match:
    user_id, date, action = match.groups()
```

**⭐ 5. Detailed Example**

```python
logs = [
    "2025-12-16 10:00 user_id=101 action=login",
    "2025-12-16 10:05 user_id=102 action=logout",
]

import re

pattern = re.compile(r"user_id=(\d+) action=(\w+)")

for log in logs:
    match = pattern.search(log)
    if match:
        user_id, action = match.groups()
        print(user_id, action)

101 login
102 logout
```

⭐ 8. Time & Space Complexity

Time Complexity: O(n * m) where n = number of lines, m = length of each line.
Space Complexity: O(k) for storing extracted groups, k = total matches.

⭐ 9. Common Pitfalls & Mistakes

❌ Using overly greedy patterns, missing specific matches.
❌ Forgetting to escape special characters (. ? * etc.)
❌ Re-compiling patterns inside a loop instead of once.
✔ Compile patterns once if reused.
✔ Use raw strings (r"pattern") to avoid escape issues.
✔ Test regex on representative text before production.
