**1. What This Pattern Solves**

Count errors

Count duplicates

Count events by category

Build fact aggregations

Top-K problems

Daily user counts

CDC row-change counts

Used constantly in logs, JSON, ETL pipelines, and Silver-layer aggregations.

**2. SQL Equivalent**

```python
%sql
SELECT key, COUNT(*) AS freq
FROM source_table
GROUP BY key;
```

**3. Core Idea**

Use a dictionary to map item → count

Increment as data flows in

Single pass, no sorting required

4. Template Code (MEMORIZE THIS)

```python
freq = {}
for x in arr:
    freq[x] = freq.get(x, 0) + 1
```

5. Detailed Example

```python
## You have log events:

events = ["200","200","404","500","200"]

## Apply template:

freq = {}
for code in events:
    freq[code] = freq.get(code, 0) + 1

## Output:
# {"200": 3, "404": 1, "500": 1}
```

```python
events = ["login", "logout", "login", "purchase", "login"]

freq = {}

for e in events:
    freq[e] = freq.get(e, 0) + 1
    
{
    "login": 3,
    "logout": 1,
    "purchase": 1
}
```

**7. Full DE Problem**

```python
## Given a list of API logs:

logs = [
  {"user":"a", "status":200},
  {"user":"b", "status":200},
  {"user":"a", "status":500},
  {"user":"c", "status":200},
  {"user":"b", "status":500}
]

## Count how many times each status code appears.
## Solution skeleton:
freq = {}
for r in logs:
    code = r["status"]
    freq[code] = freq.get(code, 0) + 1
```

```python
## You ingest clickstream events and need event-type counts per batch.

{
    "page_view": 12034,
    "add_to_cart": 3421,
    "checkout": 892
}

freq = {}
for event in events:
    etype = event["type"]
    freq[etype] = freq.get(etype, 0) + 1
```

**8. Complexity**

Time: O(n)

Space: O(k) where k = unique keys

**9. Pitfalls**

❌ Using nested loops → O(n²)

❌ Sorting before counting (unnecessary cost)

❌ Forgetting default initialization

✔ Use dict.get() or defaultdict(int) for clean increments
